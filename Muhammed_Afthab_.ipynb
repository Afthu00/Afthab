6. Do min max scaling on the feature set (Take ‘Survived’ as target)

````python
from sklearn.preprocessing import MinMaxScaler


non_numeric_cols = ['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked']


numeric_features = df.drop(columns=['Survived'] + non_numeric_cols)


scaler = MinMaxScaler()

scaled_features = scaler.fit_transform(numeric_features)


scaled_df = pd.DataFrame(scaled_features, columns=numeric_features.columns)

scaled_df['Survived'] = df['Survived']


print(scaled_df.head())
````
